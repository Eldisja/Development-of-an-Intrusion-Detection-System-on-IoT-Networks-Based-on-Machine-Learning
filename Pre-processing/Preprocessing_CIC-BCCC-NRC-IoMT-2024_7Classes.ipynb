```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
import function
```

```python
# Load original dataset
cicot2023_7classes = '/content/drive/MyDrive/Dataset_Skripsi/7_Classes/CIC-BCCC-NRC-IoMT-2024_7Classes.csv'
df = pd.read_csv(cicot2023_7classes)
```

# Cek Distribusi Kelas

```python
# Misal kolom target bernama 'label', kamu bisa pisahkan X dan y seperti ini:
X = df.drop(columns=['Attack Name'])
y = df['Attack Name']

#untuk selanjutnya sebelum encoding
y_label = y.copy()  # Simpan sebelum encoding

# tujuan agar tahu setiap angka pada label encoding mewakili apa saja.
```

```python
# Label Encoding target (Attack_type)
label_encoder_y = LabelEncoder()
y_encoded = label_encoder_y.fit_transform(y)
```

```python
# Tampilkan hasil label encoding berdampingan
label_mapping_df = pd.DataFrame({'Label_Original': y_label, 'Label_Encoded': y_encoded})
print("===== Mapping Label Encoding (Attack_type) =====")
print(label_mapping_df.drop_duplicates().sort_values(by='Label_Encoded').reset_index(drop=True).to_string(index=False))
```

```python
# Panggil function distribusi kelas
function.plot_class_distribution(y_label, y_encoded, title="Distribution of Attack Types CIC-BCCC-NRC-IoMT-2024_7Classes")
```
