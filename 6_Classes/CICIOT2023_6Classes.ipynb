```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
import functions
```

```python
# Mengambil dataset CIC IoT 2023
cic_iot2023 = '/content/drive/MyDrive/Dataset_Skripsi/Dataset_Asli/CIC IOT 2023_Merged10.csv' # download yang merged10
df = pd.read_csv(cic_iot2023)
```

```python
# Memisahkan kolom fitur dan target
X = df.drop(columns=['Label'])
y = df['Label']

# Menyimpan label asli sebelum encoding agar tidak ikut termodifikasi
y_label = y.copy()

# Melakukan encoding label utama (y) sekali untuk StratifiedKFold
y_encoded = LabelEncoder().fit_transform(y.squeeze())
```

```python
# Menghitung jumlah sampel per kelas
class_counts = y.value_counts().reset_index()
class_counts.columns = ['Label_Original', 'Jumlah_Sampel']

# Mengurutkan label bedasarkan jumlah sampel terbanyak
class_counts = class_counts.sort_values(by='Jumlah_Sampel', ascending=False).reset_index(drop=True)

print("===== Distribusi Jumlah Sampel per Kelas =====")
print(class_counts.to_string(index=False))
```

```python
# Membuat daftar label yang ingin disimpan
classes_to_keep = [
    'VULNERABILITYSCAN',
    'DDOS-ACK_FRAGMENTATION',
    'DDOS-UDP_FRAGMENTATION',
    'DNS_SPOOFING',
    'DOS-HTTP_FLOOD',
    'RECON-HOSTDISCOVERY'
]
```

```python
# Menyimpan satu baris penuh
df= df[df['Label'].isin(classes_to_keep)]

# Mengacak setiap baris
df_final = df.sample(frac=1, random_state=42).reset_index(drop=True)
```

```python
# Menyimpan ke file dataset baru
output_path = '/content/drive/MyDrive/Dataset_Skripsi/6_Classes/CIC_IOT_2023_6Classes.csv'
df_final.to_csv(output_path, index=False)

print(f"Dataset baru berhasil dibuat dan disimpan di {output_path}")
```
