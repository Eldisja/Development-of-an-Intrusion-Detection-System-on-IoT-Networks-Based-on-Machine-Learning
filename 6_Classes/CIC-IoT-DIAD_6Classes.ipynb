```python
import pandas as pd

# Membuat daftar file beserta jenis serangannya
file_label_pairs = [
    ('/content/drive/MyDrive/Dataset_Skripsi/CIC IoT-DIAD 2024/DDoS-HTTP_Flood-.pcap_Flow.csv', 'DDoS-HTTP_Flood'),
    ('/content/drive/MyDrive/Dataset_Skripsi/CIC IoT-DIAD 2024/DoS-HTTP_Flood.pcap_Flow.csv', 'DoS-HTTP_Flood'),
    ('/content/drive/MyDrive/Dataset_Skripsi/CIC IoT-DIAD 2024/DoS-UDP_Flood4.pcap_Flow.csv', 'DoS-UDP_Flood'),
    ('/content/drive/MyDrive/Dataset_Skripsi/CIC IoT-DIAD 2024/Mirai-greeth_flood1.pcap_Flow.csv', 'Mirai-greeth_flood'),
    ('/content/drive/MyDrive/Dataset_Skripsi/CIC IoT-DIAD 2024/XSS.pcap_Flow.csv','XSS'),
    ('/content/drive/MyDrive/Dataset_Skripsi/CIC IoT-DIAD 2024/SqlInjection.pcap_Flow.csv','SqlInjection')
]

```

```python
# Menampilkan jumlah sampel tiap dataset
for path, label in file_label_pairs:
    try:
        df = pd.read_csv(path)
        print(f"{label:20}: {df.shape[0]:>7} baris")
    except Exception as e:
        print(f"{label:20}: Gagal dibaca ({e})")
```

```python
# Menentukan jumlah sampel yang akan diambil dari masing-masing dataset
target_samples = [11720, 13513, 12794, 5856, 3377, 6603]

# Inisialisasi list untuk menyimpan dataframe yang telah diolah
dataframes = []
```

```python
# Mengambil dan mengupdate label pada masing-masing dataframe
for path, label in file_label_pairs:
    df = pd.read_csv(path)

    if 'Label' in df.columns:
        df['Label'] = label  # Ganti semua keterangan di kolom Label
    else:
        df['Label'] = label  # Jika kolom Label tidak ada

    dataframes.append(df)
```

```python
# Inisialisasi list hasil balancing
balanced_dataframes = []

# Melakukan sampling dan penggabungan
for i, (df, target) in enumerate(zip(dataframes, target_samples)):
    if df.shape[0] > target:
        df_sampled = df.sample(n=target, random_state=42)
    else:
        df_sampled = df.copy()

    balanced_dataframes.append(df_sampled)

    print(f"Dataset {i+1} ({file_label_pairs[i][1]}): Dipakai {df_sampled.shape[0]} sampel")

# Menggabungkan semua dataset menjadi satu dataset yang baru
df_merged = pd.concat(balanced_dataframes, ignore_index=True)
```

```python
# Menyimpan dataset ke file dataset CIC IoT DIAD
df_merged.to_csv('/content/drive/MyDrive/Dataset_Skripsi/6_Classes/CIC-IoT-IDAD_6Classes.csv', index=False)
```
